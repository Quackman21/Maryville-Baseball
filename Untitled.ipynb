```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
csv_file = 'Metrics.csv'  # Replace with the actual path to your CSV file
try:
    df = pd.read_csv(csv_file)
except FileNotFoundError:
    print(f"Error: File '{csv_file}' not found.")
    exit(1)
except pd.errors.EmptyDataError:
    print(f"Error: File '{csv_file}' is empty.")
    exit(1)
except pd.errors.ParserError:
    print(f"Error: Unable to parse '{csv_file}' as a CSV file.")
    exit(1)

# Extract the relevant columns
bat_speeds = df['Bat Speed (mph)']
rotational_accelerations = df['Rotational Acceleration (g)']
names = df['Last Name']

# Create a scatter plot
plt.figure(figsize=(12, 8))
plt.scatter(bat_speeds, rotational_accelerations, alpha=0.7, c='b', edgecolors='k', s=100)

# Add labels to data points
for i in range(len(names)):
    plt.annotate(names[i], (bat_speeds[i], rotational_accelerations[i]), fontsize=10)

# Set axis labels and title
plt.xlabel('Bat Speed (mph)')
plt.ylabel('Rotational Acceleration (g)')
plt.title('Scatter Plot of Bat Speed vs. Rotational Acceleration')

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file into a DataFrame
csv_file = 'Metrics.csv'  # Replace with the actual path to your CSV file
df = pd.read_csv(csv_file)

# Extract the relevant columns
on_plane_efficiency = df['On Plane Efficiency (%)']
vertical_bat_angle = df['Vertical Bat Angle (deg)']
names = df['Last Name']

# Create a scatter plot with a trendline
plt.figure(figsize=(12, 8))
sns.regplot(x=on_plane_efficiency, y=vertical_bat_angle, scatter=True, ci=None, label='Data', color='b', scatter_kws={"s": 100})

# Add labels to data points
for i in range(len(names)):
    plt.annotate(names[i], (on_plane_efficiency[i], vertical_bat_angle[i]), fontsize=10)

# Set axis labels and title
plt.xlabel('On Plane Efficiency (%)')
plt.ylabel('Vertical Bat Angle (deg)')
plt.title('Scatter Plot of On Plane Efficiency vs. Vertical Bat Angle with Trendline')

# Show the plot
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

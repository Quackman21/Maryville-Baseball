```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# List of players
players = [
    "Colin Bergmann", "Brady Kindhart", "Michael Long", 
    "Elias Stevens", "Matthew Arnold", "Jack Zebig",
    "Josh Herget", "Michael Gould", "Harry Oden",
    "Camden Ruby", "Steven Schnieder", "Dominic Sharp",
    "Jordan Rosenberg", "Aaris Stolte", "Jacob Hager",
    "Nick Daugherty", "Gavin Schacher", "Andy Bunton",
    "Matt Schuler", "Jordan Stucki"
]

# Replace this section with your actual data loading
all_data = pd.DataFrame({
    'Player': np.random.choice(players, 200),
    'Rotation Score': np.random.randint(20, 81, 200),
    'Plane Score': np.random.randint(20, 81, 200),
    'Connection Score': np.random.randint(20, 81, 200),
})

# Calculate average scores for each player and team average
avg_scores = all_data.groupby('Player').mean()
team_avg = avg_scores.mean()

# Adding the team average to the DataFrame
avg_scores.loc['Team Average'] = team_avg

# Sorting players by total average score
avg_scores['Total'] = avg_scores.mean(axis=1)
avg_scores = avg_scores.sort_values(by='Total', ascending=True)

# Plotting
fig, ax = plt.subplots(figsize=(12, 12))

# Create stacked bar chart
labels = list(avg_scores.index)
rotation_scores = list(avg_scores['Rotation Score'])
plane_scores = list(avg_scores['Plane Score'])
connection_scores = list(avg_scores['Connection Score'])

# Get y-axis positions
y_positions = np.arange(len(labels))

# Plot data with thinner bars
ax.barh(y_positions, rotation_scores, color='red', label='Rotation Score', height=0.4)
ax.barh(y_positions, plane_scores, left=rotation_scores, color='yellow', label='Plane Score', height=0.4)
ax.barh(y_positions, connection_scores, left=np.array(rotation_scores) + np.array(plane_scores), color='orange', label='Connection Score', height=0.4)

# Annotate each bar
for i, (rot, pln, con) in enumerate(zip(rotation_scores, plane_scores, connection_scores)):
    ax.text(rot / 2, i, str(round(rot, 2)), va='center', ha='center', color='black', weight='bold' if rot > 60 else 'normal')
    ax.text(rot + pln / 2, i, str(round(pln, 2)), va='center', ha='center', color='black', weight='bold' if pln > 60 else 'normal')
    ax.text(rot + pln + con / 2, i, str(round(con, 2)), va='center', ha='center', color='black', weight='bold' if con > 60 else 'normal')

# Add some labels and title
ax.set_yticks(y_positions)
ax.set_yticklabels(labels)
ax.set_xlabel('Scores')
ax.set_title('Player and Team Average Scores')

# Bold the "Team Average" label on the y-axis
tick_labels = ax.get_yticklabels()
tick_labels[-1].set_fontweight('bold')

# Move the legend off the graph
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Start the x-axis at 20
plt.xlim([20, None])

plt.tight_layout(rect=[0, 0, 0.85, 1])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Provided player names
player_names = [
    "Colin Bergmann", "Brady Kindhart", "Michael Long", 
    "Elias Stevens", "Matthew Arnold", "Jack Zebig",
    "Josh Herget", "Michael Gould", "Harry Oden",
    "Camden Ruby", "Steven Schnieder", "Dominic Sharp",
    "Jordan Rosenberg", "Aaris Stolte", "Jacob Hager",
    "Nick Daugherty", "Gavin Schacher", "Andy Bunton",
    "Matt Schuler", "Jordan Stucki"
]

# Initialize an empty DataFrame to store all data
all_data = pd.DataFrame()

# Load data from each sheet into the DataFrame
for name in player_names:
    df = pd.read_excel('All hitting metrics.xlsx', sheet_name=name, skiprows=range(0, 8), engine='openpyxl')
    df['Player'] = name  # Add a column to identify the player
    all_data = all_data.append(df, ignore_index=True)

# Create the plot
fig, ax = plt.subplots(figsize=(15, 10))

# Initialize empty lists to store max and average bat speeds
max_speeds = []
avg_speeds = []
sorted_player_names = []

# Calculate max and average bat speeds for each player
for player in player_names:
    player_data = all_data[all_data['Player'] == player]
    max_speed = player_data['Bat Speed (mph)'].max()
    avg_speed = player_data['Bat Speed (mph)'].mean()
    
    max_speeds.append(max_speed)
    avg_speeds.append(avg_speed)
    sorted_player_names.append(player)

# Sort players by their max bat speed
sorted_indices = sorted(range(len(max_speeds)), key=lambda k: max_speeds[k], reverse=True)
sorted_player_names = [sorted_player_names[i] for i in sorted_indices]
max_speeds = [max_speeds[i] for i in sorted_indices]
avg_speeds = [avg_speeds[i] for i in sorted_indices]

# Plotting
x = np.arange(len(sorted_player_names))
width = 0.35  # the width of the bars

ax.bar(x - width/2, max_speeds, width, label='Max Bat Speed', color='blue')
ax.bar(x + width/2, avg_speeds, width, label='Average Bat Speed', color='green')

# Adding labels
for i, (max_speed, avg_speed) in enumerate(zip(max_speeds, avg_speeds)):
    ax.text(i - width/2, max_speed + 0.5, f"{max_speed:.2f}", ha='center')
    ax.text(i + width/2, avg_speed + 0.5, f"{avg_speed:.2f}", ha='center')

# Add details to the plot
ax.set_ylabel('Bat Speed (mph)')
ax.set_xlabel('Players')
ax.set_xticks(x)
ax.set_xticklabels(sorted_player_names, rotation=45, ha="right")
ax.set_title('Max and Average Bat Speed for Each Player')

# Add a legend
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))

plt.tight_layout(rect=[0, 0, 0.85, 1])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Provided player names
player_names = [
    "Colin Bergmann", "Brady Kindhart", "Michael Long", 
    "Elias Stevens", "Matthew Arnold", "Jack Zebig",
    "Josh Herget", "Michael Gould", "Harry Oden",
    "Camden Ruby", "Steven Schnieder", "Dominic Sharp",
    "Jordan Rosenberg", "Aaris Stolte", "Jacob Hager",
    "Nick Daugherty", "Gavin Schacher", "Andy Bunton",
    "Matt Schuler", "Jordan Stucki"
]

# Players grouped by grade
seniors = ["Colin Bergmann", "Elias Stevens", "Matthew Arnold", "Michael Gould", "Jacob Hager", "Andy Bunton", "Matt Schuler", "Michael Long"]
juniors = ["Josh Herget", "Camden Ruby"]
sophomores = ["Jordan Rosenberg", "Aaris Stolte", "Nick Daugherty", "Jordan Stucki", "Jack Zebig", "Harry Oden"]
freshmen = list(set(player_names) - set(seniors) - set(juniors) - set(sophomores))

# Initialize an empty DataFrame to store all data
all_data = pd.DataFrame()

# Load data from each sheet into the DataFrame
for name in player_names:
    df = pd.read_excel('All hitting metrics.xlsx', sheet_name=name, skiprows=range(0, 8), engine='openpyxl')
    df['Player'] = name  # Add a column to identify the player
    all_data = all_data.append(df, ignore_index=True)

# Function to create a bar chart for a specific grade
def create_bar_chart(players, grade):
    max_speeds = []
    avg_speeds = []

    for player in players:
        player_data = all_data[all_data['Player'] == player]
        max_speed = player_data['Bat Speed (mph)'].max()
        avg_speed = player_data['Bat Speed (mph)'].mean()
        max_speeds.append(max_speed)
        avg_speeds.append(avg_speed)

    x = np.arange(len(players))
    width = 0.35

    fig, ax = plt.subplots(figsize=(12, 8))

    ax.bar(x - width/2, max_speeds, width, label='Max Bat Speed', color='blue')
    ax.bar(x + width/2, avg_speeds, width, label='Average Bat Speed', color='green')

    for i, (max_speed, avg_speed) in enumerate(zip(max_speeds, avg_speeds)):
        ax.text(i - width/2, max_speed + 0.5, f"{max_speed:.2f}", ha='center')
        ax.text(i + width/2, avg_speed + 0.5, f"{avg_speed:.2f}", ha='center')

    ax.set_ylabel('Bat Speed (mph)')
    ax.set_title(f'{grade} Players: Max and Average Bat Speed')
    ax.set_xticks(x)
    ax.set_xticklabels(players, rotation=45)
    ax.legend(loc='upper left', bbox_to_anchor=(1, 1))

    plt.tight_layout(rect=[0, 0, 0.85, 1])
    plt.show()

# Create bar charts for each grade
create_bar_chart(seniors, 'Senior')
create_bar_chart(juniors, 'Junior')
create_bar_chart(sophomores, 'Sophomore')
create_bar_chart(freshmen, 'Freshman')

```
